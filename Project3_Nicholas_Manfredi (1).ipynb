```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np


def randGraph(V, E):
    A = np.zeros((V, V))
    edges_so_far = 0
    while edges_so_far < E:
        i = randint(0, V - 1)
        j = randint(0, V - 1)
        if (i != j and A[i, j] == 0):
            A[i, j] = 1
            A[j, i] = 1
            edges_so_far += 1
    return A

A = randGraph(10, 9)
G = nx.Graph(A)
nx.draw(G)
```
