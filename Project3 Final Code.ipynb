```python
import networkx as nx
import numpy as np

def randomGraph(n, p):
    A = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            if i < j:
                if random() < p:
                    A[i, j] = 1
                    A[j, i] = 1
    return A

def connected_c(A):
    global visited
    visited = [False] * len(A)
    comp = []
    for i in range(len(A)):
        if visited[i] == False:
            comp.append(component(A, i))
    return comp

def component(A, i):
    c = {i}
    visited[i] = True
    for j in range(len(A)):
        if visited[j] == False and A[i, j] == 1:
            c|=component(A, j)
    return c 
    
A = randomGraph(50, 0.01)
%time connected_c(A)
```

```python
def connected(A): ##question four (optimized code)
    global visited
    visited = [False] * len(A)
    comp = []
    for i in range(len(A)):
        if visited[i] == False:
            comp.append(len(component(A, i)))
    comp.sort(reverse = True) 
    return comp[0]

A = randomGraph(1000, 1/1000)
%time connected(A)
```

```python
def randomGraph2(n, p):
    L = [[] for i in range(n)]
    for i in range(n):
        for j in range(i + 1, n):
            if random() < p:
                L[i].append(j)
                L[j].append(i)
    return L

def component2(L, i):
    c = {i}
    visited[i] = True
    for j in L[i]:
        if not visited[i]:
            c|= component2(L, j)
    return c

def connected2(A): 
    global visited
    visited = [False] * len(A)
    comp = []
    for i in range(len(A)):
        if visited[i] == False:
            comp.append(len(component2(A, i)))
    comp.sort(reverse = True)
    return comp[0]

```
